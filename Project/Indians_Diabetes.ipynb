#### Load dữ liệu

```python
from numpy import loadtxt
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

Bộ dữ liệu gồm có $9$ phần tử trên mỗi hàng, $8$ phần tử đầu là các biến đầu vào $X$ và biến đầu ra $y$ nhận giá trị $\{0,1\}$

Bài toán đặt ra là dự báo việc có khởi phát bệnh tiểu đường trong $5$ năm tới. 
8 giá trị đầu trong mỗi hàng được diễn giải dưới đây
1. Số lần mang thai
2. Nồng độ glucose trong huyết tương sau $2$ giờ trong xét nghiệm dung nạp glucose đường uống.
3. Huyết áp tâm trương (mm Hg)
4. Độ dày của nếp gấp da ở cánh tay (cơ tam đầu) giữa vai và đùi (mm)
5. insulin huyết thanh trong 2 giờ (mu U/ml)
6. Chỉ số BMI (kg / m^2)
7. Tỷ lệ di truyền bệnh tiểu đường
8. Tuổi (năm)

Biến đầu ra $y$ nhận giá trị $0$ hoặc $1$

```python
# Load dữ liệu
dataset = loadtxt('pima_indians_diabetes.csv', delimiter=',')
# Chia dữ liệu
X = dataset[:,0:8]
y = dataset[:,8]
```

#### Tạo mô hình mạng Neuron trong Keras 

```python
model = Sequential()
# layer đầu tiên trong hidden layer gồm 12 nodes, hàm kích hoạt là hàm ReLU
model.add(Dense(12, input_shape=(8,), activation='relu'))
# layer thứ hai trong hidden layer gồm 8 nodes, hàm kích hoạt là hàm ReLU
model.add(Dense(8, activation='relu'))
# layer thứ 3 chỉ có một node và hàm kích hoạt là hàm sigmoid
model.add(Dense(1, activation='sigmoid'))
```

#### Điều chỉnh mô hình

```python
# compile the keras model
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
```

#### Chạy mô hình

```python
print(X.shape)
print(y.shape)
```

```python
# fit the keras model on the dataset
model.fit(X, y, epochs=150, batch_size=10)
```

```python
accuracy = model.evaluate(X, y)
print("Accuracy: %.2f" % accuracy[-1]) 
```

##### Dự đoán

```python
# make class predictions with the model
predictions = (model.predict(X) > 0.5).astype(int)
# summarize the first 5 cases
for i in range(5):
    print('%s => %d (expected %d)' % (X[i].tolist(), predictions[i], y[i]))
```
